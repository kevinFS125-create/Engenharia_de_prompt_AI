Missão 01: Automação de Arquivos

```python
def organizar_por_extensao(pasta):
    # Percorre todos os arquivos da pasta informada
    for arquivo in os.listdir(pasta):
        caminho_arquivo = os.path.join(pasta, arquivo)

        # Verifica se o item encontrado é realmente um arquivo (não uma pasta)
        if os.path.isfile(caminho_arquivo):
            # Pega a extensão do arquivo (parte depois do ponto)
            ext = arquivo.split(".")[-1]

            # Cria uma pasta com o nome da extensão (em maiúsculo), se não existir
            pasta_destino = os.path.join(pasta, ext.upper())
            os.makedirs(pasta_destino, exist_ok=True)

            # Move o arquivo para dentro da pasta correspondente à sua extensão
            shutil.move(caminho_arquivo, os.path.join(pasta_destino, arquivo))

    # Mensagem final indicando que a organização foi concluída
    print("Arquivos organizados por extensão!")
```
